```python
# Import dependencies
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.plot(x_axis, y_axis, label='Ninjo Messin')
plt.title("General Messin")
plt.xlabel('Duration of Jo')
plt.ylabel('Rate of Messin')
plt.legend()
plt.grid()

plt.savefig("analysis/NinjoMessin.png")

plt.show()
```

```python
# Object oriented interface method: Create the plot with ax.plt()
#fig = plt.figure()
#ax = fig.add_subplot()

# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)

plt.show()
```

```python
# Object oriented interface method: Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()

ax.plot(x_axis, y_axis)

plt.show()
```

```python
# Also Objected oriented interface method: Quick Graph3
#Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis)

#To change figure-level attributes (such as axis labels, a title, or a legend) or save the figure as an image, use fig = plt.figure().
```
